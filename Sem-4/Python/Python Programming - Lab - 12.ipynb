### 07) WAP to demonstrate the use of Histogram.

```python
import random
cpis = [random.randint(0,10) for _ in range(100)]
m.hist(cpis,bins=10,histtype='stepfilled',align = 'mid',label = 'CPI Hist')
m.legend()
m.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = [1,2,3,4,5]
y = [25,50,75,100,125]
c = ['g','b','c','y','r']
m.barh(x,y,color = c)

for index, value in enumerate(y):
    m.text(value, index,
             str(value))

m.show()
```
